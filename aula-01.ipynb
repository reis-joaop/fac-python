### Variáveis
Antes de continuarmos, precisamos falar sobre **variáveis**. Pense em uma variável como uma 'caixa' onde você pode guardar um valor para usar depois. Cada variável precisa de um nome.

```python
nome = "João" # Guardando o texto 'João' na variável chamada 'nome'
idade = 25    # Guardando o número 25 na variável chamada 'idade'

# Podemos imprimir o conteúdo dessas variáveis
print(nome)
print(idade)
```

```python
name = "Reis" # A hashtag é usada para fazer comentarios em códigos python
age = 19

print(f"O aluno {name} tem {age} anos de idade")
```

## 2. Entrada de Dados com `input()`

Enquanto o `print()` joga informações para fora, o `input()` pega informações que vêm de fora (do usuário) para dentro do programa.
Quando o Python lê um `input()`, ele pausa o programa e espera você digitar algo e apertar Enter.

```python
# Vai aparecer uma caixa de texto pedindo o nome
seu_nome = input("Digite o seu nome: ")

print(f"Olá, {seu_nome}! Prazer em te conhecer.")
```

```python
your_name = input("Type your name: ")

print(f"Hello, {your_name}! Nice to meet you.")
```

**Atenção:** Tudo o que o `input()` recebe é tratado como **texto** (string), mesmo que você digite um número! Se quisermos fazer contas matemáticas, precisaremos converter esse texto para número.

## 3. Operações Matemáticas Básicas

Python é uma excelente calculadora! Veja os operadores básicos:

*   `+` : Adição
*   `-` : Subtração
*   `*` : Multiplicação
*   `/` : Divisão (resultado com casas decimais)
*   `//`: Divisão Inteira (resultado sem casas decimais)
*   `%` : Resto da divisão (Módulo)
*   `**`: Exponenciação (Potência)

```python
a = 10
b = 3

print(f"Soma: {a + b}")
print(f"Subtração: {a - b}")
print(f"Multiplicação: {a * b}")
print(f"Divisão: {a / b}")
print(f"Divisão Inteira: {a // b}")
print(f"Resto da Divisão: {a % b}")
print(f"Potência (10 elevado a 3): {a ** b}")
```

```python
a = 28
b = 3

print(f"{a} plus {b} is: {a+b}")
print(f"{a} minus {b} is: {a-b}")
print(f"{a} divided by {b} is: {a/b}")
print(f"{a} divided by {b} (with no decimals) is: {a//b}")
print(f"What remains of {a} divided by {b} is: {a % b}")
print(f"{a} cubed is: {a ** b}")
```

## 4. Juntando Tudo: Funções Matemáticas + Entrada/Saída

Vamos fazer um programa que pede dois números ao usuário e calcula a soma. Como dissemos antes, o `input()` retorna um texto (string). Precisamos usar a função `int()` (para números inteiros) ou `float()` (para números com vírgula) para converter o texto em número.

```python
# Recebendo os números como texto (input) e logo em seguida convertendo para número inteiro (int)
numero1 = int(input("Digite o primeiro número inteiro: "))
numero2 = int(input("Digite o segundo número inteiro: "))

soma = numero1 + numero2

print(f"A soma de {numero1} com {numero2} é igual a {soma}.")
```

```python
num1 = int(input("Type here your first number: "))
num2 = int(input("Type here your seconde number:"))

print(f"The sum of {num1} plus {num2} its equal to {num1 + num2}")
```

### Funções Matemáticas Mais Usadas (Módulo `math` e padrões)

Além dos operadores básicos, Python possui funções prontas para ajudar. Algumas já estão embutidas (`abs()`, `round()`), enquanto outras precisam ser chamadas de uma biblioteca matemática chamada `math`.

```python
import math

numero_negativo = -15
numero_quebrado = 3.74159

print(f"Valor Absoluto (tira o sinal): abs(-15) = {abs(numero_negativo)}")
print(f"Arredondamento: round(3.74159, 2 casas) = {round(numero_quebrado, 2)}")
print(f"Arredondamento para Cima: math.ceil(3.74159) = {math.ceil(numero_quebrado)}")
print(f"Arredondamento para Baixo: math.floor(3.74159) = {math.floor(numero_quebrado)}")
print(f"Raiz Quadrada de 9: math.sqrt(9) = {math.sqrt(9)}")
```

```python
import math

negative_number = -30
decimal_number = 9.94212
print(f"Absolute Value (makes the number positive, despite the number being negative): abs(-30) = {abs(negative_number)}")
print(f"Roundening the number: round(9.94212, 3 two decimal place) = {round(decimal_number, 3)}")
print(f"Roundening up the number: math.ceil(9.94212) = {math.ceil(decimal_number)}")
print(f"Roundening down the number: math.floor(9.94212 = {math.floor(decimal_number)})")
print(f"The product of which number times itself is 9: math.sqrt(9) = {math.sqrt(9)}")
```

## 5. Gerador de Números Aleatórios (Sorteios)

Muitas vezes, em jogos ou sorteios, precisamos que o computador escolha um número aleatório (sortido) por nós. O Python tem uma biblioteca interna chamada `random` exatamente para isso.

```python
import random

# Sorteando um número inteiro entre 1 e 10 (incluindo o 10)
numero_sorteado = random.randint(1, 10)
print(f"O número sorteado de 1 a 10 foi: {numero_sorteado}")

# Escolhendo um elemento aleatório de uma lista de opções (veremos listas mais a frente, mas é útil saber!)
opcoes = ["Cara", "Coroa"]
moeda = random.choice(opcoes)
print(f"Joguei a moeda e deu: {moeda}")
```

```python
import random
# to make the computer choose within a pre planned range use "radint" from random library
random_number = random.randint(0, 10)
print(f"The random numer this time is: {random_number}")

# to choose an element within a list use "choice"
random_options = ["Heads", "Tails"]
coin = random.choice(random_options)
print(f"I toss a coin, its: {coin}")
```
